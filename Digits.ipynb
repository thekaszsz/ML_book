```python
from sklearn.model_selection import train_test_split

X = digits.data
y = digits.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
print('train =',len(X_train))
print('test =',len(X_test))
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score

model = svm.SVC(kernel = 'rbf', gamma = 0.001)
model.fit(X_train, y_train)

pred = model.predict(X_test)
score = accuracy_score(y_test, pred)
print('ì •ë‹µë¥  :', score*100, '%')
```

```python
from sklearn.datasets import load_digits
from sklearn import decomposition
from mpl_toolkits.mplot3d import Axes3D

digits = load_digits()
X = digits.data
y = digits.target

# 0ã€œ9ì˜ ìƒ‰ ì´ë¦„ì„ ì¤€ë¹„í•œë‹¤
numbercolor = ["BLACK","BROWN","RED","DARKORANGE","GOLD",
               "GREEN","BLUE","PURPLE","GRAY","SKYBLUE"]
# y ê°’ì„ ìƒ‰ ì´ë¦„ìœ¼ë¡œ ë°”ê¾¸ê³  colors ë¦¬ìŠ¤íŠ¸ë¥¼ ë§Œë“ ë‹¤.
colors = []
for i in y:
	colors.append(numbercolor[i])

# ì£¼ì„±ë¶„ ë¶„ì„ìœ¼ë¡œ 64ê°œì˜ íŠ¹ì§•ëŸ‰ì„ 3ê°œë¡œ ì°¨ì›ì„ ì¶•ì†Œí•œë‹¤
pca = decomposition.PCA(n_components=3)
features3 = pca.fit_transform(X)

# 3ê°œë¡œ ì¤„ì¸ ë°ì´í„°(features3)ë¡œ ë°ì´í„° í”„ë ˆìž„ì„ ë§Œë“ ë‹¤
df = pd.DataFrame(features3)

# 3D ì‚°í¬ë„ë¥¼ ì¤€ë¹„í•œë‹¤
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
# 3ê°œì˜ íŠ¹ì§•ëŸ‰ì„ X,Y,Zë¡œ í•˜ê³ , ê° ì ì˜ ìˆ«ìžì— ëŒ€ì‘í•˜ëŠ” ìƒ‰ìœ¼ë¡œ ì‚°í¬ë„ë¥¼ ê·¸ë¦°ë‹¤
ax.scatter(df[0], df[1], df[2], color=colors)

# ìˆ«ìžê°€ ê°ê° ë¬´ìŠ¨ ìƒ‰ì¸ì§€ ê²¬ë³¸ì„ ê·¸ë¦°ë‹¤
ty = 0
for col in numbercolor:
    ax.text(50, 30, 30-ty*5, str(ty), size=20, color=col)
    ty+=1
plt.show()
```
