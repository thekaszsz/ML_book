총 네 가지 특징량 데이터가 들어 있음

```python
import pandas as pd
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.head()
```

iris를 DataFrame에 넣고 잘 나왔는지 앞 5행만 보기

```python
df.columns = iris.feature_names
```

열 이름 설정

```python
df["target"] = iris.target
```

target을 열로 추가

```python
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

#세 가지 품종을 별개의 df로 나눔
#target 값이 0이라면 df0에, 1이라면 df1에, 2라면 df2에 넣는다
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]

#세 가지 품종을 히스토그램에서 색으로 구분해 그린다
#꽃받침의 폭으로 히스토그램을 그린다
plt.figure(figsize = (5,5))
xx = "sepal width (cm)" #sepal width (cm) 열을 대상으로
df0[xx].hist(color = "b", alpha = 0.5) #파란색 히스토그램 작성
df1[xx].hist(color = "r", alpha = 0.5) #빨간색 히스토그램 작성
df2[xx].hist(color = "g", alpha = 0.5) #초록색 히스토그램 작성

plt.xlabel(xx)
plt.show()
```

xx = 꽃받침 폭이라는 특징량 한 개만 선택하여 히스토그램을 그려 봄

꽃받침 폭을 기점으로 본다면, 품종의 차이가 있다는 것을 볼 수 있다
하지만 겹치는 부분이 많아서 확실한 구별이 어렵다

두 개의 특징량 사용 feat.산포도

```python
#꽃받침 폭과 길이로 산포도 그리기
xx = "sepal width (cm)"
yy = "sepal length (cm)"
plt.figure(figsize=(5,5))
plt.scatter(df0[xx], df0[yy].color="b", alpha = 0.5) #파란색 산포도 작성
plt.scatter(df1[xx], df0[yy].color="r", alpha = 0.5) #빨간색 산포도 작성
plt.scatter(df2[xx], df0[yy].color="g", alpha = 0.5) #초록색 산포도 작성

plt.xlabel(xx)
plt.ylabel(yy)
plt.show()
```

python 버전 차이로 인해 책에서는 .color="b"로 되어 있었으나, 내 버전에서는 , c="b"로 고쳐야 함

```python
#꽃받침 폭과 길이로 산포도 그리기
xx = "sepal width (cm)"
yy = "sepal length (cm)"
plt.figure(figsize=(5,5))
plt.scatter(df0[xx], df0[yy], c="b", alpha = 0.5) #파란색 산포도 작성
plt.scatter(df1[xx], df0[yy], c="r", alpha = 0.5) #빨간색 산포도 작성
plt.scatter(df2[xx], df0[yy], c="g", alpha = 0.5) #초록색 산포도 작성

plt.xlabel(xx)
plt.ylabel(yy)
plt.grid()
plt.show()
```

책에서는 파랑/빨강+초록으로 구별 지을 수 있다고 하나, 내가 그린 산포도에서는 구별 짓기에는 다소 어려움이 있다

3D 산포도로 다시 그릴 예정

```python
from mpl_toolkits.mplot3d import Axes3D
#꽃받침 폭, 꽃받침 길이, 꽃잎 길이로 3D 산포도를 그릴 예정
xx = "sepal width (cm)"
yy = "sepal length (cm)"
zz = "petal length (cm)"

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection = "3d")
ax.scatter(df0[xx], df0[yy], df0[zz], color = "b")
ax.scatter(df1[xx], df1[yy], df1[zz], color = "r")
ax.scatter(df2[xx], df2[yy], df2[zz], color = "g")

ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
plt.show
```

2차원보다는 3차원이 훨씬 더 구별이 쉬워짐 그러나 아직도 빨간색과 초록색이 구별되기 어려워 '시점'을 바꿀 예정

```python
fig = plt.figure(figsize =(5,5))
ax = fig.add_subplot(projection = "3d")

ax.scatter(df0[xx], df0[yy], df0[zz], color = "b")
ax.scatter(df1[xx], df1[yy], df1[zz], color = "r")
ax.scatter(df2[xx], df2[yy], df2[zz], color = "g")

ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
ax.view_init(0,240) #시점을 바꾼다
plt.show
```

시점을 바꾸면 해결할 수 없을 것 같은 문제도 해결책을 찾을 수 있다

```python
fig = plt.figure(figsize =(5,5))
ax = fig.add_subplot(projection = "3d")

ax.scatter(df0[xx], df0[yy], df0[zz], color = "b")
ax.scatter(df1[xx], df1[yy], df1[zz], color = "r")
ax.scatter(df2[xx], df2[yy], df2[zz], color = "g")

ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
ax.view_init(25, 90) #시점을 바꾼다
plt.show
```
