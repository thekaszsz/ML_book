```python
import pandas as pd
import numpy as np
```

```python
emp = pd.read_csv('/content/drive/MyDrive/data100/emp.csv')
```

```python
emp['hiredate'] = pd.to_datetime(emp['hiredate'])
emp.hiredate.dt.year
```

```python
emp.info()
```

```python
emp['hire_year'] = emp.hiredate.dt.year
```

```python
emp.groupby('hire_year')['sal'].mean().reset_index()
```

```python
result = emp.groupby('hire_year')['sal'].mean().reset_index()
result.columns=['년도','평균월급']
result.sort_values(by = '평균월급', ascending = False)
```

```python
emp.groupby('job')['empno'].count().reset_index()
```

```python
result = emp.groupby('job')['empno'].count().reset_index()
result.columns = ['직업','인원수']
result[:][result['인원수'] >= 3]
```

```python
emp.groupby(['deptno','job'])['sal'].sum().reset_index()
```

```python
emp['순위'] = emp['sal'].rank(ascending=False).astype(int)
emp[['ename','sal','순위']].sort_values(by = '순위',ascending = True)
```

```python
emp20 = pd.read_csv('/content/drive/MyDrive/data100/emp20.csv', encoding = 'euckr')
```

```python
emp20['순위'] = emp20['age'].rank(method = 'dense', ascending=False).astype(int)
emp20[['ename','age','순위']].sort_values(by = '순위', ascending = True)
```

```python
result = emp[:][emp['job'] == 'SALESMAN']
result['순위'] = result['sal'].rank(method = 'dense', ascending = False).astype(int)
result[['ename','sal','job','순위']][result['job'] == 'SALESMAN'].sort_values(by = '순위', ascending = True)
```

```python
emp['순위'] = emp.groupby('deptno')['sal'].rank(ascending=False).astype(int)
emp[['deptno','ename','sal','순위']].sort_values(by = ['deptno', '순위'])
```

```python
emp20['순위'] = emp20.groupby('telecom')['age'].rank(ascending = False).astype(int)
emp20[['telecom','ename','age','순위']].sort_values(by = ['telecom','순위'])
```

```python
emp['순위'] = emp['sal'].rank(method= 'dense', ascending=False).astype(int)
cuts = 4
emp['sal_grade'] = pd.qcut(emp.순위, q = cuts, labels = range(1,cuts+1))
emp[['ename','sal','sal_grade']].sort_values(by = 'sal_grade', ascending = True)
```

```python
emp20['순위'] = emp20['age'].rank(method = 'dense', ascending=False).astype(int)
cuts = 5
emp20['age_grade'] = pd.qcut( emp20.순위, q = cuts, labels = range(1,cuts+1))
emp20[['ename','age','age_grade']].sort_values(by = 'age_grade')
```
